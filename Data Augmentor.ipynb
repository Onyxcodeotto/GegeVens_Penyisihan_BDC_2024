```
thesaurus_url = "https://raw.githubusercontent.com/victoriasovereigne/tesaurus/master/dict.json"
thesaurus_response = requests.get(thesaurus_url)
```

```
thesaurus = thesaurus_response.json()
thesaurus
```

```
thesaurus_tags = tags = {entry['tag'] for entry in thesaurus.values()}
```

```
def select_synonyms(word, thesaurus):
    if word in thesaurus:
        return np.random.choice(thesaurus[word]["sinonim"])
    else:
        return None

def create_synonym_text(array_of_text, thesaurus, synonym_chance):
    array_copy = deepcopy(array_of_text)
    for i in range(len(array_of_text)):
        print(array_copy[i])
        if synonym_chance > np.random.random():
            array_copy[i] = select_synonyms(array_of_text[i], thesaurus)
    return array_copy
```
